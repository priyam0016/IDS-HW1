```python
#List is a heterogenous collection 
#Example List in Python
myList = [1, 2, 3, 4]
print(myList)

#Accessing the elements in the list
print(myList[0])

#Changing the value at some Index
myList[1]=10
print(myList)

#Checking the element if contained in the list
if 3 in myList:
  print ('Element found')

#Length of a list
print(myList.__len__())

#Iteration through list
for x in myList:
  print(x, end='|')

#Appending a element to the list
print("\nAppending a element to the list")
myList.append(5)
print(myList)

#Index of the element in the  list
print(myList.index(3))  #Output will be 2

#Insert a element in the list at some index
myList.insert(3,7)
print(myList.__str__())

#Sorting the above list
myList.sort()   #Gives the list in ascending order
myList.sort(reverse=True) #Gives the list in descending order
print(myList)


#hetrogenity of collections
print("hetrogeneous list")
myList2 = [1, "A", "Anuj", True] #In Python, list can contain any type of values, numbers, strings, etc.
print(myList2)

#pop out the last added element from list as in Stack
print("pop out the last added element from list")
print(myList2.pop())  #Print/ pop out TRUE

#Reversing the list
print("Reverse list is printed")
myList2.reverse()
print (myList2)

#Concatanating two lists
print("Concatanating two lists")
myList3 = myList + myList2
print(myList3)

#Counting number of occurrances of a element in the list
print("Counting number of occurrances of a element in the list")
print(myList3.count(1)) #Output will be 2 since LIst 3 created after concatanating has 1 two times

#Second way to concatanate two list using Extend
print("Concatanation using extend")
myList.extend(myList2)
print(myList)

#Clearing all elements in the list
myList.clear() 

print("Size reduces to 0 after clearing all elements in the lsit")
print(myList.__len__())
```
